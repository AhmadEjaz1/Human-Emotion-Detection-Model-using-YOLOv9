```python
from ultralytics import YOLO

# Load the model
model = YOLO('yolov9s.pt')  # Use 'yolov8s.pt' for a small model

# Train the model
results = model.train(
    data="/content/DataSet_1/data.yaml",  # Path to your dataset configuration file
    epochs=100,               # Number of epochs
    batch=16,                 # Batch size
    imgsz=640,                # Image size
    save=True,                # Save checkpoints
    device=0,                 # Use GPU
    lr0=0.01,                 # Initial learning rate
    optimizer='auto',         # Automatically select the best optimizer
    augment=True,             # Enable data augmentation
    patience=50,              # Early stopping if no improvement for 50 epochs
)
```

```python
#Code for Removing extra cordinates from label files to match the yolo label formats
import os

label_dir = "/content/DataSet_1/DataSet/labels/train"
for label_file in os.listdir(label_dir):
    with open(os.path.join(label_dir, label_file), 'r') as f:
        lines = f.readlines()

    # Keep only the first 5 values in each line
    cleaned_lines = []
    for line in lines:
        values = line.strip().split()
        if len(values) >= 5:
            cleaned_lines.append(" ".join(values[:5]) + "\n")

    # Write the cleaned lines back to the file
    with open(os.path.join(label_dir, label_file), 'w') as f:
        f.writelines(cleaned_lines)
```
